```python
import torch
from transformers import pipeline
```

```python
pipe = pipeline(
    "image-text-to-text",
    model = "google/gemma-3-4b-it",
    device = "cuda",
    torch_dtype=torch.bfloat16,
)
```

```python
messages = [
    {"role":"user",
    "content":[
        {
            "type":"image",
            "url":"https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/datasets/cat_ip_adapter.png"
        },
        {
            "type":"text",
            "text":"which animals is in the picture and can you decribe it?"
        }
    ]}
]
```

```python
output = pipe(text=messages, max_new_tokens = 200)
```

```python
print(output[0]["generated_text"][-1]["content"])
```
