```python
from transformers import pipeline
```

## Text Classification

```python
text = "This movie is beautiful. I would like to watch this movie again"
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier(text)
```

## Zero Shot Classification

```python
classifier = pipeline("zero-shot-classification")
```

```python
text = "This is a tutorial about Hugging Face"
```

```python
candidate_labels = ["tech", "education", "business"]
```

```python
classifier(text, candidate_labels)
```

## Text Generation

```python
generator = pipeline("text-generation")
```

```python
prompt = "This tutorial will walk you through how to"
```

```python
generator(prompt, max_length = 30)
```

```python
generator = pipeline("text-generation", model = "distilgpt2")
```

```python
generator(prompt, max_length = 30)
```

## Named Entity Recognition (NER)

```python
ner = pipeline("ner", grouped_entities = True)
```

```python
sentence = "My name is Tirendaz and I work Google Cloud in Berlin"
```

```python
ner(sentence)
```

```python
question_answerer = pipeline("question-answering")
```

```python
question_answerer(
    question = "Where do I work?",
    context = sentence
)
```
