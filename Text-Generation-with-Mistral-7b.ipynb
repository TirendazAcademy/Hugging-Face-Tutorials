# Setting the Quantization

```python
from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_4bit = True,
    bnb_4bit_quanty_type = "nf4",
    bnb_4bit_use_double_quanty = True
)
```

# Model Loading

```python
from transformers import AutoTokenizer, AutoModelForCausalLM 
import torch

model_name = "/kaggle/input/mistral/pytorch/7b-v0.1-hf/1"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_4bit = True,
    quantization_config = bnb_config,
    torch_dtype = torch.bfloat16,
    device_map = "auto",
    trust_remote_code = True
)
```

# Creating the Pipeline

```python
from transformers import pipeline 

pipe = pipeline(
    "text-generation",
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.bfloat16,
    device_map = "auto"
)
```

# Text-Generation

```python
prompt = "As a data scientist, can you explain what is overfitting?"

sequence = pipe(
    prompt,
    do_sample = True,
    top_k = 50,
    top_p = 0.95,
    max_new_tokens = 100,
    temperature = 0.7,
    num_return_sequences = 1,
)

print(sequence[0]["generated_text"])
```
